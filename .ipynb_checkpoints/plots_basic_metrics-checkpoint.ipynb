```python
from os.path import exists as exists
import pandas as pd
```

```python
RULES = [1,2]
enlaces = [1,2,4,8,16,32]
experimentos = list(range(1,11))
path_plot_files = "plot_files\\"
degradation_type = ["attack", "failure"]
```

```python
def save_csv(data_df, path_file):
    print("Save: ", path_file)
    data_df.to_csv(path_file, sep=',',encoding='utf-8')
```

# Obtiene datos de degradaci√≥n (ataques o fallos)

```python
def get_df_degradation():
    PATH_BASE = dict()
    data_attack = dict()
    for rule in rules:
        PATH_BASE[rule] = dict() 
        data_attack[rule] = dict() 
        for enlace in enlaces:
            PATH_BASE[rule][enlace] = dict()
            data_attack[rule][enlace] = dict() 
            for experimento in experimentos:
                p_base = "graphs-degradation\\r" + str(rule) + "\\enlaceD" + str(enlace) + "\\experimento" + str(experimento) + "\\"
                if exists(p_base):
                    PATH_BASE[rule][enlace][experimento] = p_base
                    p_data_attack = p_base + 'ataques\\1\\degradationData.csv'
                    df_data_attack = pd.read_csv(p_data_attack, index_col='#n_nodes')
                    df_data_attack.index.names = ['num_nodes']
                    print(df_data_attack)
                    data_attack[rule][enlace][experimento] = df_data_attack
                else:
                    print("No existe: ", p_base)
    return data_attack
```

Plots files para calcular metricas

```python
def get_data_attack(metrics):
    data = dict()
    degradation_data = get_df_attack()
    for metric in metrics:
        data[metric] = dict()
        for enlace in enlaces:
            data[metric][enlace] = dict() 
            for rule in rules:
                data[metric][enlace][rule] = dict() 
                df_experiments = list()
                for experimento in experimentos:
                    if metric == 'prop_largestc':
                        # entre el orden del grafo original
                        #df = degradation_data[rule][enlace][experimento][['o_largestc']].div(2500)
                        # entre el orden del grafo con fallos
                        df = degradation_data[rule][enlace][experimento][['o_largestc']].div(df.index, axis=0)
                        df = df.rename(columns={"o_largestc": metric})
                    elif metric == 'avg_isolatedc':
                        df = degradation_data[rule][enlace][experimento][['o_largestc']].sub(df.index, axis=0)
                        df = df.rename(columns={"o_largestc": metric})
                        df = df * -1
                        df_n_comp = degradation_data[rule][enlace][experimento][['n_comp']] - 1
                        print(df_n_comp)
                        df = df[[metric]].div(df_n_comp.n_comp, axis=0)
                        df = df.fillna(0)
                    else:
                        df = degradation_data[rule][enlace][experimento][[metric]]
                    df = df.assign(experiment = experimento)
                    #print(df)
                    if not df.empty:
                        df_experiments.append(df)
                    else:
                        print("No existe el archivo " + p_data)
                data[metric][enlace][rule] = pd.concat(df_experiments)
    return data
```

# Crea archivos de plots

```python
def compute_plot_files(metrics_plots, data_metrics, degradation):
    results = dict()
    for metric in metrics_plots:
        results[metric] = dict()
        for enlace in enlaces:
            results[metric][enlace] = dict()
            p_results = path_plot_files + metric + '\\' + metric + '_enlaceD' + str(enlace) + '_' + degradation + '.csv'
            if not exists(p_results):
                print(p_results)
                results[metric][enlace]['removed'] = [x/100 for x in list(range(100,-1,-1))]
                if metric == 'components':
                    results[metric][enlace]['mean_r1_s'] = data_metrics['prop_largestc'][enlace][1].groupby('num_nodes')['prop_largestc'].mean()
                    results[metric][enlace]['std_r1_s'] = data_metrics['prop_largestc'][enlace][1].groupby('num_nodes')['prop_largestc'].std()
                    results[metric][enlace]['mean_r2_s'] = data_metrics['prop_largestc'][enlace][2].groupby('num_nodes')['prop_largestc'].mean()
                    results[metric][enlace]['std_r2_s'] = data_metrics['prop_largestc'][enlace][2].groupby('num_nodes')['prop_largestc'].std()
                    results[metric][enlace]['mean_r1_avg_s'] = data_metrics['avg_isolatedc'][enlace][1].groupby('num_nodes')['avg_isolatedc'].mean()
                    results[metric][enlace]['std_r1_avg_s'] = data_metrics['avg_isolatedc'][enlace][1].groupby('num_nodes')['avg_isolatedc'].std()
                    results[metric][enlace]['mean_r2_avg_s'] = data_metrics['avg_isolatedc'][enlace][2].groupby('num_nodes')['avg_isolatedc'].mean()
                    results[metric][enlace]['std_r2_avg_s'] = data_metrics['avg_isolatedc'][enlace][2].groupby('num_nodes')['avg_isolatedc'].std()
                    #results[metric][enlace]['mean_r1_avg_s'][results[metric][enlace]['mean_r1_avg_s'].eq(0)] = ''
                    #results[metric][enlace]['std_r1_avg_s'][results[metric][enlace]['std_r1_avg_s'].eq(0)] = ''
                    #results[metric][enlace]['mean_r2_avg_s'][results[metric][enlace]['mean_r2_avg_s'].eq(0)] = ''
                    #results[metric][enlace]['std_r2_avg_s'][results[metric][enlace]['std_r2_avg_s'].eq(0)] = ''
                else:
                    results[metric][enlace]['mean_r1'] = data_metrics[metric][enlace][1].groupby('num_nodes')[metric].mean()
                    results[metric][enlace]['std_r1'] = data_metrics[metric][enlace][1].groupby('num_nodes')[metric].std()
                    results[metric][enlace]['mean_r2'] = data_metrics[metric][enlace][2].groupby('num_nodes')[metric].mean()
                    results[metric][enlace]['std_r2'] = data_metrics[metric][enlace][2].groupby('num_nodes')[metric].std()
                pd_results = pd.DataFrame(results[metric][enlace])
                pd_results.index.name = 'num_nodes'
                pd_results = pd_results.fillna(0)
                pd_results = pd_results.sort_values(by=['num_nodes'],ascending=False)
                print(pd_results)
                save_csv(pd_results, p_results)
            else:
                print('Ya existe ', p_results)
```
