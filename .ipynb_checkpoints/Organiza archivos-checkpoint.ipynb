```python
import os
```

```python
# Cambia el nombre de las secuencias de ataques
for r in [1,2]:
    for l in [1,2,4,8,16,32]:
        for e in range(1,11):
            path = 'graphs-degradation/r' + str(r) + '/enlaceD' + str(l) + '/experimento' + str(e) + '/ataques/1/AttackSequence.txt'
            if os.path.exists(path):
                new_name = 'new_files/r' + str(r) + '_lmaxL' + str(l) +  '_e' + str(e) + '_nodeAttackSequence.txt'
                print(new_name)
                print(path)  
                os.rename(path, new_name)
```
