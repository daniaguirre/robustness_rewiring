```python
import pandas as pd
import matplotlib.pyplot as plt
```

En los siguientes bloques se calculan los datos de degradación de un experimento de FALLO de R1 y y otro en R2 solo para comparar
- *df1*: contiene los datos de la secuencia 1 de fallos del experimento 1, enlace D1 y regla R1
- *df2*: contiene los datos de la secuencia 1 de fallos del experimento 1, enlace D1 y regla R2

```python
df1 = pd.read_csv("r1/enlaceD1/experimento1/fallos/1/degradationData.csv")
df2 = pd.read_csv("r2/enlaceD1/experimento1/fallos/1/degradationData.csv")

# Datos para el ejemplo de r1
df1["f"] = df1.index                                    # fracción de nodos en fallos
df1["fslc"] = df1["o_largestc"]/2500                    # fslc = Número de nodos en LC/Número de nodos en el grafo original
df1["o_relativo"] = df1["o_largestc"]/df1["#n_nodes"]   # orden relativo = Número de nodos en LC/Número de nodos aún vivos
df1.head(5)

# Datos para el ejemplo de r2
df2["f"] = df2.index                                    # fracción de nodos en fallos
df2["fslc"] = df2["o_largestc"]/2500                    # fslc = Número de nodos en LC/Número de nodos en el grafo original
df2["o_relativo"] = df2["o_largestc"]/df2["#n_nodes"]   # orden relativo = Número de nodos en LC/Número de nodos aún vivos

df2.head(3)
```

Ahora hacemos lo mismo pero para comparar un ejemplo de la degradación de ATAQUES en r1 contra uno de r2

```python
da1 = pd.read_csv("r1/enlaceD1/experimento1/ataques/1/degradationData.csv")
da2 = pd.read_csv("r2/enlaceD1/experimento1/ataques/1/degradationData.csv")
da1["f"] = da1.index
da1["fslc"] = da1["o_largestc"]/2500
da1["o_relativo"] = da1["o_largestc"]/da1["#n_nodes"]
da2["f"] = da2.index
da2["fslc"] = da2["o_largestc"]/2500
da2["o_relativo"] = da2["o_largestc"]/da2["#n_nodes"]
da2.head(3)
```

A continuación se compara el *fslc* (entre $S_{LC}(0)$) vs *orden relativo* (entre $S_{LC}(f)$) dentro del mismo experimento

```python
# Gráfica de fallos con R1, experimento 1, secuencia 1

plt.scatter(x=df1["f"], y=df1["fslc"], alpha=0.5, marker="^")
plt.scatter(x=df1["f"], y=df1["o_relativo"], alpha=0.5)
plt.legend(["$S_{LC}(f)~/~S_{LC}(0)$", "$S_{LC}(f)~/~S_{G}(f)$"])
plt.xlabel("f (percentage of nodes removed)")
plt.title("Failures R1")
custom_dashes = (2, 2)
plt.grid(linestyle='--', color='gray',alpha=0.2,linewidth=1)
plt.show()
```

```python

# Gráfica de fallos con R2, experimento 1, secuencia 1
plt.scatter(x = df2["f"], y = df2["fslc"], alpha=0.5, marker="^")
plt.scatter(x = df2["f"], y = df2["o_relativo"], alpha=0.5)
plt.legend(["$S_{LC}(f)~/~S_{LC}(0)$", "$S_{LC}(f)~/~S_{G}(f)$"])
plt.xlabel("f (percentage of nodes removed)")
plt.title("Failures R2")
custom_dashes = (2, 2)
plt.grid(linestyle='--', color='gray',alpha=0.2,linewidth=1)
plt.show()
```

Ahora comparamos fslc vs orden letativo dentro del mismo experimento para ataques

```python
# Gráfica de ataques con R1, experimento 1, secuencia 1

plt.scatter(x=da1["f"], y=da1["fslc"], alpha=0.5, marker="^")
plt.scatter(x=da1["f"], y=da1["o_relativo"], alpha=0.5)
plt.legend(["$S_{LC}(f)~/~S_{LC}(0)$", "$S_{LC}(f)~/~S_{G}(f)$"])
plt.xlabel("f (percentage of nodes removed)")
plt.title("Attacks R1")
custom_dashes = (2, 2)
plt.grid(linestyle='--', color='gray',alpha=0.2,linewidth=1)
plt.show()
```

Las siguientes gráficas comparan el ejemplo de r1 vs r2

# Gráficas de puntos críticos

A continuación leemos los puntos críticos pci, pcm y pcf de todos los experimento calculados con respecto al orden del componente más grande inicial:$S_{LC}(0)$

```python
pc_f_r1_s0 = pd.read_csv("r1/puntos_criticos_r1_fallos_s0.csv")
pc_f_r2_s0 = pd.read_csv("r2/puntos_criticos_r2_fallos_S0.csv")
pc_a_r1_s0 = pd.read_csv("r1/puntos_criticos_r1_ataques_S0.csv")
pc_a_r2_s0 = pd.read_csv("r2/puntos_criticos_r2_ataques_S0.csv")
```

A continuación leemos los puntos críticos pci, pcm y pcf de todos los experimento calculados con respecto al orden del componente más grande inicial:$S_{LC}(f)$

```python
pc_f_r1_sf = pd.read_csv("r1/puntos_criticos_r1_fallos_sf.csv")
pc_f_r2_sf = pd.read_csv("r2/puntos_criticos_r2_fallos_Sf.csv")
pc_a_r1_sf = pd.read_csv("r1/puntos_criticos_r1_ataques_Sf.csv")
pc_a_r2_sf = pd.read_csv("r2/puntos_criticos_r2_ataques_Sf.csv")
```

```python

#promedios_f_r1_s0 = pc_f_r1_s0.groupby('enlace').mean()
promedios_f_r1_s0 = pc_f_r1_s0.groupby('enlace').agg({
    'pci': 'mean',
    'pcm': 'mean',
    'pcf': 'mean',
}).reset_index()
std_f_r1_s0 = pc_f_r1_s0.groupby('enlace').std()
promedios_f_r1_s0

```

```python
promedios_f_r2_s0 = pc_f_r2_s0.groupby('enlace').agg({
    'pci': 'mean',
    'pcm': 'mean',
    'pcf': 'mean',
}).reset_index()
std_f_r2_s0 = pc_f_r2_s0.groupby('enlace').std()
promedios_f_r2_s0
```

```python

# Configura el tamaño del gráfico
#promedios_f_r1_s0['enlace'] = pd.Categorical(promedios_f_r1_s0['enlace'])
error_config = {'ecolor': 'black', 'capsize': 4, 'capthick': 1, 'elinewidth': 1}
# Grafica las tres series
plt.errorbar(promedios_f_r1_s0['enlace'], promedios_f_r1_s0['pci'], yerr=std_f_r1_s0['pci'], label='$cp_i$', alpha=0.5, marker='o', **error_config)
plt.errorbar(promedios_f_r1_s0['enlace'], promedios_f_r1_s0['pcm'], yerr=std_f_r1_s0['pcm'], label='$cp_m$', alpha=0.5, marker='s', **error_config)
plt.errorbar(promedios_f_r1_s0['enlace'], promedios_f_r1_s0['pcf'], yerr=std_f_r1_s0['pcf'], label='$cp_f$', alpha=0.5, marker='^', **error_config)

# Personaliza el gráfico
plt.xlabel('$l_{max}$')
plt.ylabel('$f$ (percentage of nodes removed)')
plt.ylim(0, 100)
plt.title('Critical points of r1 failure degradation $S_{LC}(f)~/~S_{LC}(0)$')

# Agrega una leyenda
plt.legend()

# Muestra el gráfico
plt.grid(linestyle='--', color='gray',alpha=0.2,linewidth=1)  # Agrega una cuadrícula al gráfico
plt.show()

# NOTA: no supe cómo se verían mejor las leyendas así que las dejé en español, tal vez tu encuentres una manera mejor de escribirlas
```
